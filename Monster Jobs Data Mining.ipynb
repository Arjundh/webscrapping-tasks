```python
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.common.by import By
from selenium.webdriver.support.expected_conditions import  visibility_of_element_located
from selenium.webdriver.common.action_chains import ActionChains
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.chrome.options import Options
import re
import pandas as pd
import numpy as np
import time
```

```python
#launch url
all_section=[]
driver = webdriver.Firefox(executable_path='C://Users//Arjun D H//geckodriver-v0.24.0-win64//geckodriver.exe')
for index in range(0,5):
    print(index)
    url ="https://www.monsterindia.com/search/walkin-jobs"+str(index)+"0"
    driver.get(url)
    driver.implicitly_wait(50)
    section = driver.find_elements_by_xpath("//*[@class='card-apply-content']")
    for i in section:
        all_section.append(i.text)
driver.close()
```

```python
print(len(all_section))
```

```python
print(all_section[0])
```

```python
import pandas as pd
df=pd.DataFrame(index=range(0,125),columns=['job_Role','company','location','experience','description'])
for idx,section in enumerate(all_section):
    scn=section.split('\n')
    try:
        df.iloc[idx,0]=scn[0] #job_role
    except IndexError:
        df.iloc[idx,0]=np.nan
    try:
        df.iloc[idx,1]=scn[1] #company
    except IndexError:
        df.iloc[idx,1]=np.nan
    try:
        df.iloc[idx,2]=scn[2] #location
    except IndexError:
        df.iloc[idx,2]=np.nan
    try:
        df.iloc[idx,3]=scn[3] #experience
    except IndexError:
        df.iloc[idx,3]=np.nan
    try:
        df.iloc[idx,4]=scn[4] #description
    except IndexError:
        df.iloc[idx,4]=np.nan
        
```
